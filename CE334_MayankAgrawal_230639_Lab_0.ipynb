```python
#Q13
import numpy as np

def diff_val(func, point, method='central', h=0.01):       # checking which method to use
    if method == 'forward':
        d = (func(point + h) - func(point)) / h

    elif method == 'backward':
        d = (func(point) - func(point - h)) / h

    else:                                                    # central difference (default)
        d = (func(point + h) - func(point - h)) / (2 * h)

    return d

#example check

x0 = 0.5
res1 = diff_val(np.exp, x0)
print("Derivative of e^x at x =", x0, ":", res1)

print()

res2 = diff_val(np.cos, x0)
print("Derivative of cos(x) at x =", x0, ":", res2)
```

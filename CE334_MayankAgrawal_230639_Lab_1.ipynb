#Q1 part (a)
Converted latitude, longitude and height into Cartesian coordinates using WGS-84 values. We found the radius of curvature, then calculated the x, y and z coordinates for Kanpur and my hometown.

```python
#Q1 part (a)
import numpy as np
import math
# phi = latitude
# lam = longitude
# h = elevation

def geo_to_xyz(lat_deg, lon_deg, ht):
  a_val =  6378137
  f_inv = 298.257223563
  f_val = 1 / f_inv

  ecc_sq = 2 * f_val - f_val * f_val

  lat_r = math.radians(lat_deg)        #conversion to radian
  lon_r = math.radians(lon_deg)

  sin_lat = math.sin(lat_r)             #calculation of N value
  temp = 1 - ecc_sq * (sin_lat*sin_lat)
  N_rad = a_val / math.sqrt(temp)

  Xc = (N_rad + ht) * math.cos(lat_r) * math.cos(lon_r)     #conversion into cartesian coordinates
  Yc = (N_rad + ht) * math.cos(lat_r) * math.sin(lon_r)
  Zc = ((1 - ecc_sq) * N_rad + ht) * math.sin(lat_r)

  return Xc, Yc, Zc

###########################################
  # Kanpur geodetic coordinates
lat_kanpur = 26 + 26/60 + 59.7228/3600
lon_kanpur = 80 + 19/60 + 54.7356/3600
h_kanpur = 126.630                            # height(in m)

x_k, y_k, z_k = geo_to_xyz(lat_kanpur, lon_kanpur, h_kanpur)

print("Kanpur Coordinates")
print("X =", x_k)
print("Y =", y_k)
print("Z =", z_k)

print()
# My hometown is Gangapur City, Rajasthan (322201)
#Latitude: 26° 28' 18.156'' N
#Longitude: 76° 42' 57.384'' E
#Height = 260 m

lat_ggc = 26 + 28/60 + 18.156/3600        # Ganagpur City(hometown) geodetic coordinates
lon_ggc = 76 + 42/60 + 57.384/3600
h_ggc = 260                           # height(in m)

x_k, y_k, z_k = geo_to_xyz(lat_ggc, lon_ggc, h_ggc)

print("Gangapur City(Hometown) Coordinates")
print("X =", x_k)
print("Y =", y_k)
print("Z =", z_k)

```

#Q1 part (b)
cartesian xyz coordinate is converted back into geodetic form. Longitude is found directly while latitude and height are calculated using iterative method because earth is ellipsoid and not perfect sphere.

```python
#Q1 part (b)
import math

def xyz_to_geo(Xv, Yv, Zv):
    A = 6378137.0
    inv_f = 298.257223563
    f1 = 1.0 / inv_f

    e_sq = 2*f1 - (f1*f1)       # eccentricity square

    lon_val = math.atan2(Yv, Xv)   # longitude directly

    p_xy = math.sqrt(Xv*Xv + Yv*Yv)     # distance in xy plane

    lat_old = math.atan((Zv / p_xy) / (1 - e_sq))    # taking lattitude phi0

    #iteration
    for _ in range(10):                 # usually converges very fast
        sin_l = math.sin(lat_old)

        N_now = A / math.sqrt(1 - e_sq*(sin_l*sin_l))    # radius of curvature

        h_now = (p_xy / math.cos(lat_old)) - N_now         # update the height

        lat_new = math.atan(                          # new latitude
            (Zv / p_xy) /
            (1 - (e_sq * N_now) / (N_now + h_now))
        )

        if abs(lat_new - lat_old) < 1e-8:           #taking tolerance of order (10^-12)
            break                                   # break if change is tiny

        lat_old = lat_new

    lat_deg = math.degrees(lat_new)               # final outputs in degrees
    lon_deg = math.degrees(lon_val)

    return lat_deg, lon_deg, h_now      #return
```

#Q2 part(a)
cartesian coordinate is converted into spherical form. radius is calculated from x y z and then latitude and longitude is found using inverse trigonometric formula.

```python
#Q2 part(a)
# phi = latitude
# lam = longitude

# cartesian to spherical
def cart_to_sph(xp, yp, zp):
    r_out = math.sqrt(xp*xp + yp*yp + zp*zp)       # radius from origin

    phi_out = math.asin(zp / r_out)                # latitude
    lam_out = math.atan2(yp, xp)                   # longitude

    phi_deg = math.degrees(phi_out)           # convert back to degrees
    lam_deg = math.degrees(lam_out)

    return r_out, phi_deg, lam_deg

#kanpur coordinates check
Xp = 959683.7651469298
Yp = 5633246.91869837
Zp = 2823834.838180068

r_out, phi_out, lam_out = cart_to_sph(Xp, Yp, Zp)
print("Kanpur Coordinates")
print("Radius   :", r_out)
print("Latitude :", phi_out)
print("Longitude:", lam_out)
```

#Q2 part(b)
Spherical coordinate is converted into cartesian using radius, latitude and longitude. Angles are first change into radian and then simple formula is used to find x, y and z value.

```python
#Q2 part(b)
# phi = latitude
# lam = longitude

# spherical to cartesian
def sph_to_cart(r_val, phi_deg, lam_deg):

    phi_r = math.radians(phi_deg)             # angle to radian
    lam_r = math.radians(lam_deg)

    x_c = r_val * math.cos(phi_r) * math.cos(lam_r)
    y_c = r_val * math.cos(phi_r) * math.sin(lam_r)
    z_c = r_val * math.sin(phi_r)

    return x_c, y_c, z_c

#kanpur coordinates check
r = 6374049.495370327
lat = 26.29676276302304
longi = 80.33187099999999

x_c, y_c, z_c = sph_to_cart(r, lat, longi)
print("Kanpur Coordinates")
print("X =", x_c)
print("Y =", y_c)
print("Z =", z_c)
```

#Q3
Geodetic coordinate is first converted into cartesian using ellipsoid parameter, then cartesian coordinate is change into spherical form by finding radius and angles. Same function is used for Kanpur and Gangapur city example.

```python
#Q3

import math

def geo_to_spher(lat_d, lon_d, hgt,
                     a_ref=6378137.0,
                     inv_flat=298.257223563,
                     sphere_rad=None):
    # if sphere radius not given, take it same as ellipsoid semi major axis
    if sphere_rad is None:
        sphere_rad = a_ref

    f_val = 1.0 / inv_flat    #flattning

    e2_val = 2*f_val - f_val*f_val      # eccentricity sq.

    # angle to radian
    lat_r = math.radians(lat_d)
    lon_r = math.radians(lon_d)

    # step 1: geodetic to cartesian
    sin_lat = math.sin(lat_r)
    N_cur = a_ref / math.sqrt(1 - e2_val*(sin_lat*sin_lat))

    x_p = (N_cur + hgt) * math.cos(lat_r) * math.cos(lon_r)
    y_p = (N_cur + hgt) * math.cos(lat_r) * math.sin(lon_r)
    z_p = ((1 - e2_val)*N_cur + hgt) * math.sin(lat_r)

    # step 2: cartesian to spherical
    r_val = math.sqrt(x_p*x_p + y_p*y_p + z_p*z_p)
    phi_s = math.asin(z_p / r_val)
    lam_s = math.atan2(y_p, x_p)

    # convert spherical angles back to degrees
    phi_deg = math.degrees(phi_s)
    lam_deg = math.degrees(lam_s)

    return phi_deg, lam_deg, r_val

#######################################
#Kanpur example

lat_k = 26 + 26/60 + 59.7228/3600
lon_k = 80 + 19/60 + 54.7356/3600
h_k = 126.630

phi_s, lam_s, r_s = geo_to_spher(lat_k, lon_k, h_k)
print("Kanpur Coordinates")
print("Spherical Latitude  :", phi_s)
print("Spherical Longitude :", lam_s)
print("Radius (r)          :", r_s)

#######################################
print()
lat_ggc = 26 + 28/60 + 18.156/3600        # Ganagpur City(hometown) geodetic coordinates
lon_ggc = 76 + 42/60 + 57.384/3600
h_ggc = 260                           # height(in m)

phi_k, lam_k, r_k = geo_to_spher(lat_ggc, lon_ggc, h_ggc)
print("Gangapur City(Hometown) Coordinates")
print("Spherical Latitude  :", phi_k)
print("Spherical Longitude :", lam_k)
print("Radius (r)          :", r_k)
```

#Q4
First satellite and observer coordinate are converted into cartesian form, then their difference is rotated into east north up system using observer location to get topocentric coordinate.

```python
#Q4

#Kanpur coordinates -> satellite position
lat_k = 26 + 26/60 + 59.7228/3600
lon_k = 80 + 19/60 + 54.7356/3600
h_k = 126.630 + 800000                # 800 km above Kanpur

#Gangapur city(hometown) coordinates -> observer position
lat_g = 26 + 28/60 + 18.156/3600
lon_g = 76 + 42/60 + 57.384/3600
h_g = 260

import math

def geo_to_xyz(lat_d, lon_d, hgt,               # geodetic to cartesian (ECEF)
               a_val=6378137.0,
               inv_f=298.257223563):

    f_tmp = 1.0 / inv_f
    e_sq = 2*f_tmp - f_tmp*f_tmp

    lat_r = math.radians(lat_d)
    lon_r = math.radians(lon_d)

    s_lat = math.sin(lat_r)
    N_val = a_val / math.sqrt(1 - e_sq*s_lat*s_lat)

    Xv = (N_val + hgt) * math.cos(lat_r) * math.cos(lon_r)
    Yv = (N_val + hgt) * math.cos(lat_r) * math.sin(lon_r)
    Zv = ((1 - e_sq)*N_val + hgt) * math.sin(lat_r)

    return Xv, Yv, Zv


def geo_to_topo(sat_lat, sat_lon, sat_h,          # geodetic to topocentric ENU
                obs_lat, obs_lon, obs_h):

    xs, ys, zs = geo_to_xyz(sat_lat, sat_lon, sat_h)       # satellite ECEF

    xo, yo, zo = geo_to_xyz(obs_lat, obs_lon, obs_h)        # observer ECEF

    # difference vector
    dx = xs - xo
    dy = ys - yo
    dz = zs - zo

    # observer angles
    lat_o = math.radians(obs_lat)
    lon_o = math.radians(obs_lon)

    # ENU transformation
    east = -math.sin(lon_o)*dx + math.cos(lon_o)*dy
    north = (-math.sin(lat_o)*math.cos(lon_o)*dx
             -math.sin(lat_o)*math.sin(lon_o)*dy
             +math.cos(lat_o)*dz)
    up = (math.cos(lat_o)*math.cos(lon_o)*dx
          +math.cos(lat_o)*math.sin(lon_o)*dy
          +math.sin(lat_o)*dz)

    return east, north, up

E, N, U = geo_to_topo(lat_k, lon_k, h_k,
                      lat_g, lon_g, h_g)

print("Satellite topocentric coordinates relative to my hometown")
print("east(E) :", E,"m")
print("north(N) :", N, "m")
print("up   (U) :", U, "m")
```

#Bonus Q.
First hometown lat long height is change into cartesian coordinate. After that given east north up value is convert into cartesian difference and added with observer point, and in last step cartesian is again change into geodatic coordinate of aircraft.

```python
#Bonus Q.
import math

# geodetic to ECEF
def geo_to_xyz(lat_d, lon_d, hgt,
               a_val=6378137.0,
               inv_f=298.257223563):

    f_tmp = 1.0 / inv_f
    e2 = 2*f_tmp - f_tmp*f_tmp

    lat_r = math.radians(lat_d)
    lon_r = math.radians(lon_d)

    Nn = a_val / math.sqrt(1 - e2*(math.sin(lat_r)**2))

    X = (Nn + hgt) * math.cos(lat_r) * math.cos(lon_r)
    Y = (Nn + hgt) * math.cos(lat_r) * math.sin(lon_r)
    Z = ((1 - e2)*Nn + hgt) * math.sin(lat_r)

    return X, Y, Z


# ECEF to geodetic (iterative)
def xyz_to_geo(X, Y, Z,
               a_val=6378137.0,
               inv_f=298.257223563):

    f_tmp = 1.0 / inv_f
    e2 = 2*f_tmp - f_tmp*f_tmp

    lon = math.atan2(Y, X)
    p = math.sqrt(X*X + Y*Y)

    lat = math.atan(Z / (p*(1 - e2)))
    for _ in range(8):
        N = a_val / math.sqrt(1 - e2*(math.sin(lat)**2))
        h = p/math.cos(lat) - N
        lat = math.atan(Z / (p*(1 - e2*N/(N+h))))

    return math.degrees(lat), math.degrees(lon), h


# ENU to ECEF difference
def enu_to_ecef(e, n, u, lat_o, lon_o):
    lat_r = math.radians(lat_o)
    lon_r = math.radians(lon_o)

    dx = (-math.sin(lon_r)*e
          -math.sin(lat_r)*math.cos(lon_r)*n
          +math.cos(lat_r)*math.cos(lon_r)*u)

    dy = ( math.cos(lon_r)*e
          -math.sin(lat_r)*math.sin(lon_r)*n
          +math.cos(lat_r)*math.sin(lon_r)*u)

    dz = ( math.cos(lat_r)*n
          +math.sin(lat_r)*u)

    return dx, dy, dz

##################################
#data given for above functions

# Gangapur City (observer)
lat_g = 26 + 28/60 + 18.156/3600
lon_g = 76 + 42/60 + 57.384/3600
h_g = 260

# aircraft ENU
E = -800
N = 900
U = 6000

xo, yo, zo = geo_to_xyz(lat_g, lon_g, h_g)      # observer ECEF

dx, dy, dz = enu_to_ecef(E, N, U, lat_g, lon_g)  # ENU → ECEF difference

# aircraft ECEF
xa = xo + dx
ya = yo + dy
za = zo + dz

lat_a, lon_a, h_a = xyz_to_geo(xa, ya, za)     # aircraft geodetic coordinates

print("Aircraft Latitude(phi) :", lat_a)
print("Aircraft Longitude(lambda):", lon_a)
print("Aircraft Height(h)   :", h_a)
```
